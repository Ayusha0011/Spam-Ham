```python
import pandas as pd
import numpy as np

np.random.seed(42)

data = {
    "Age": np.random.randint(20, 50, 10).astype(float),
    "Salary": np.random.randint(30000, 100000, 10).astype(float),
    "Experience": np.random.randint(1, 10, 10).astype(float)
}
df = pd.DataFrame(data)

df.loc[[2,5,7],"Age"] = np.nan
df.loc[[1,4],"Salary"] = np.nan
df.loc[[3,6],"Experience"] = np.nan
print(df)
```

Replacing data with mean when data is normally distributed

```python
df_mean = df.copy()

df_mean["Age"] = df_mean["Age"].fillna(df_mean["Age"].mean())
df_mean["Salary"] = df_mean["Salary"].fillna(df_mean["Salary"].mean())
df_mean["Experience"] = df_mean["Experience"].fillna(df_mean["Experience"].mean())

print(df_mean)
```

inplace = true means modify the original dataframe directly
inplace = false mean return a new dataframe
